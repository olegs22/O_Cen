```python
from scipy.interpolate import interp1d
```

```python
def EdNdE_primary(mass,energy,col):
    a=interp1d(interp(mass,col)[0],interp(mass,col)[1])
    return a(energy)
```

```python
def phi_primary(pars,energy,col):
    mass,log_sig,log_J,log_N = pars
    
    sigmav = 10.**log_sig
    J = 10.**log_J
    N = 10.**log_N
    
    return N+J*(1./(4.*np.pi))*sigmav*EdNdE_primary(mass,energy,col)/(2.*(mass**2))
```

```python
energy = np.loadtxt('energy.dat')
Dphi = np.loadtxt('flux.dat')
err = np.loadtxt('data_errors.dat')
```

```python
plt.errorbar(energy,Dphi,yerr=err,fmt='-o',ms=3.5);
```

```python
def lnhood(pars,data,err,energy,col):
    model = phi_primary(pars,energy,col)
    
    p = (((data - model) / err)**2) + np.log(2. * np.pi * err**2)
    
    return -0.5*np.sum(p)
```

```python
def priors(pars,plist):
    mass,log_sig,log_J,log_N = pars
    
    if plist[0]<mass<plist[1] and plist[2]<log_sig<plist[3] and\
    plist[4]<log_J<plist[5] and plist[6]<log_N<plist[7]:
        return 0.0
    return -np.inf
```
